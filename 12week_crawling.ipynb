```python
from lib2to3.pgen2 import driver
import time
from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys

browser = webdriver.Chrome()
url = 'https://www.op.gg/'
browser.get(url)



#1.닉네임 입력
nick_name = input()
elem = browser.find_element_by_id('searchSummoner')
elem.send_keys(nick_name)
elem = browser.find_element_by_xpath('//*[@id="__next"]/div[2]/div[2]/div/form/button')
time.sleep(3)
asdf = elem.click()




#스크롤 내리기
#browser.execute_script('window.scrollTo(0,document.body.scrollHeight)')
interval = 1


prev_height = browser.execute_script('return document.body.scrollHeight')

while True:
    browser.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    time.sleep(interval)
    try:
        elem = browser.find_element_by_xpath("//button[@class='more'][contains(text(),'더 보기')]")
        #print(elem)
    except:
        break


    if elem:
        elem.click()

        time.sleep(interval)
        curr_height = browser.execute_script('return document.body.scrollHeight')
        
        if curr_height == prev_height:
         break

        prev_height = curr_height
    
    else:
        break
print('스크롤 완료 end')




soup = BeautifulSoup(browser.page_source, 'lxml')



games = soup.find_all("li", attrs={"class":"css-1qq23jn e1iiyghw3"})



for game in games:
    #게임 타입
    tt = game.find("div",attrs={"class":"type"}).get_text()
    #언제 했는지
    time = game.find('div',attrs={'class':'time-stamp'}).get_text()
    #승패
    res = game.find('div',attrs={'class':'result'}).get_text()
    #챔피언
    champ = game.find('img')['alt']
    #게임시간
    length = game.find('div',attrs={'class':'length'}).get_text()
    #스펠
    spells = game.find_all('div',attrs={"class":"spell"})
    spell1 = spells[0].find('img')['alt']
    spell2 = spells[1].find('img')['alt']
    #룬
    runes = game.find_all('div',attrs={"class":"rune"})
    main = runes[0].find('img')['alt']
    sub = runes[1].find('img')['alt']
    #kda
    kda = game.find('div',attrs={'class':'k-d-a'}).get_text()
    #평점
    ratio = game.find('div',attrs={'class':'ratio'}).get_text()

    #킬관여
    p_kill = game.find('div',attrs={'class':'p-kill'}).get_text()
    #제어와드
    ward = game.find('div',attrs={'class':'ward'}).get_text()
    #미니언 개수
    cs = game.find('div',attrs={'class':'cs'}).get_text()
    #평균티어
    avert = game.find('div',attrs={'class':'average-tier'})
    if avert:
        avert_text = avert.get_text()
    else:
        avert_text = '-'


    print(time + ', ' + tt+ ', ' + champ + ', ' + res+ ', ' + length + ', ' + spell1 +'/'+ spell2 +', ' + main +'/'+ sub)
    print(kda + ', ' + ratio)
    print(f'킬관여율: {p_kill}, {ward}, {cs}, 평균티어: {avert_text}')
    print('--------------------------------------------------------------------')



elem = browser.find_element_by_xpath('//*[@id="content-container"]/div[2]/ul/li[1]/div/div[2]/div[1]/div[1]/div[1]/a')
elem.click()

browser.switch_to.window(browser.window_handles[-1])

soup = BeautifulSoup(browser.page_source, 'lxml')



recent = soup.find("div",attrs={"class":"info-box"})
recent_name = recent.find("span",attrs={"class":"name"}).get_text()
recent_build = recent.find("span",attrs={"class":"build-position"}).get_text()
recent_tier = recent.find("span",attrs={"class":"tier-color"}).get_text()


print(recent_name,recent_build,recent_tier)


boxs = soup.find_all("div",attrs={"class":"css-yyrh89 ew1oorz6"})
for box in boxs:
  title = box.find("div",attrs={"class":"css-9ervy2 ew1oorz5"}).get_text()
  figure = box.find("div",attrs={"class":"css-nx8mki ew1oorz4"}).get_text()
  print(f'{title}: {figure}')
```
